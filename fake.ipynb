```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')
```

```python
data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake_manual_testing = data_fake.tail(10) #Taking the last 10 rows for manual testing
for i in range(23480, 23470,-1): #Removing the last 10 rows
    data_fake.drop([i], axis = 0, inplace = True) 
    
data_true_manual_testing = data_true.tail(10)
for i in range(21416, 21406,-1):
    data_true.drop([i], axis = 0, inplace = True)
```

```python
data_fake_manual_testing['class'] = 0
data_true_manual_testing['class'] = 1
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text): #text cleaning function
    text = text.lower() #converting text to lowercase
    text = re.sub('\[.*?\]', '', text) #removing text in square brackets
    text = re.sub("\\W"," ",text)  #removing special characters
    text = re.sub('https?://\S+|www\.\S+', '', text) #removing links
    text = re.sub('<.*?>+', '', text) #removing html tags
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) #removing punctuation
    text = re.sub('\n', '', text) #removing new line characters
    text = re.sub('\w*\d\w*', '', text) #removing words containing numbers
    return text
```

```python
data['text'] = data['text'].apply(wordopt) #applying the function to the text column
```

```python
x= data['text'] #features
y= data['class'] #labels
```

```python
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)  #splitting the data into training and testing sets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer() 
xv_train = vectorization.fit_transform(x_train) #vectorizing the training data
xv_test = vectorization.transform(x_test) #vectorizing the testing data
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()  
LR.fit(xv_train, y_train) 
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier()
GB.fit(xv_train, y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Real News"
    
def manual_testing(news):
    testing_news = {'text':[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test['text'].apply(wordopt)
    new_xv_test = vectorization.transform(new_def_test['text'])
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)
    
    return print("\n\nLogistic Regression Prediction: {} \nDecision Tree Prediction: {} \nGradient Boosting Prediction: {} \nRandom Forest Prediction: {}".format(output_lable(pred_LR[0]),output_lable(pred_DT[0]),output_lable(pred_GB[0]),output_lable(pred_RF[0])))
```
